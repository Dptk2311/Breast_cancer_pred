Label Encoding

```python
#Get Datatypes of each columns in out dataset
df.dtypes
```

```python
#Encode the diagnosis values
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
df.iloc[:,1]=labelencoder.fit_transform(df.iloc[:,1].values)
```

```python
#display df
df
```

Split Dataset & Feauture Scaling

```python
#Splitting the dataset into dependent and independent datasets
X=df.iloc[:,2:].values #X contains all the features
Y = df.iloc[:,1].astype(int).values #Y is dependent or what will be predicted
```

```python
#Splitting the dataset into training(75%) and testing(25%)
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.25)
```

```python
#Scaling the data(feauture scaling)
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.fit_transform(X_test)
```

```python
#print data
(X_train)
```

Build a Logistic Regression Model

```python
#Build a logistic Regression classifier
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train,Y_train)
```

```python
#make use of trained model to make predictions on test data
predictions = classifier.predict(X_test)
```

Performance Evaluation

```python
#plot confusion matrix
from sklearn.metrics import confusion_matrix
#visualizing confusion matrix
import seaborn as sns
cm = confusion_matrix(Y_test,predictions)
print(cm)
sns.heatmap(cm,annot=True)
```

```python
#get accuracy for model
from sklearn.metrics import accuracy_score
accuracy=(accuracy_score(Y_test,predictions))*100
print(accuracy,'%')
```
